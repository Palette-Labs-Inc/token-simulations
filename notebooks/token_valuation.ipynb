```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
## User Inputs
A_t = np.linspace(0, 1, 100)
E = np.ones(100)* 0.3

def graph_value_to_tokens(graph_value):
    return graph_value

def graph_value_to_tokens2(graph_value):
    return graph_value**0.5

def graph_value_to_tokens3(graph_value):
    return np.log1p(graph_value)

```

```python
S1_t = graph_value_to_tokens(A_t) + np.cumsum(graph_value_to_tokens(E*A_t))
S2_t = graph_value_to_tokens2(A_t) + np.cumsum(graph_value_to_tokens2(E*A_t))
S3_t = graph_value_to_tokens3(A_t) + np.cumsum(graph_value_to_tokens3(E*A_t))

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axx = ax[0]
axx.plot(A_t, label='A_t')
axx.plot(S1_t, label='M(t)=t')
axx.plot(S2_t, label='M(t)=t^0.5')
axx.plot(S3_t, label='M(t)=log(1+t)')
axx.legend()
axx.set_ylabel('S(t)')
axx.set_xlabel('t')

axx = ax[1]
axx.plot(graph_value_to_tokens(A_t)/S1_t, label='M(t)=t')
axx.plot(graph_value_to_tokens2(A_t)/S2_t, label='M(t)=t^0.5')
axx.plot(graph_value_to_tokens3(A_t)/S3_t, label='M(t)=log(1+t)')
axx.set_ylabel('Mt(At)/S(t)')
axx.set_xlabel('t')
```
